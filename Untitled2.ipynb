### Importing Libraries 

```python
import pandas as pd
import pyodbc
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Define a connection String

```python
#Define Connection parameters
Database = 'PRACTICE1'
Server = 'DESKTOP-U507UFU'
conn_str = (
    "Driver={SQL Server};"
    f"Database={Database};"
    f"Server={Server};"
    "Trusted_Connection =yes"   
)
conn = pyodbc.connect(conn_str)
```

```python
query = """
        SELECT  DISTINCT(customer_id), SUM(amount) AS Total_amount_spent FROM customer_contracts
        GROUP BY customer_id
        ORDER BY Total_amount_spent DESC
"""
df = pd.read_sql(query, conn)
```

```python
#Group Customers by Ammount Spent
#Elimiminate White Spaces from the columns
df.columns = df.columns.str.strip()
```

```python
#Group Customers by Ammount Spent
df = df.sort_values(by ='Total_amount_spent', ascending = False )
```

```python
df.head()
```

```python
Top10 = df.head(15).copy()
```

```python
#Visualize the top 15 Customers(Premium Customers)
Top10['customer_id'] = Top10['customer_id'].astype(str)
plt.figure(figsize = (10,5))
colors = sns.color_palette("Blues", len(Top10))
sns.barplot(data=Top10,  
            y = 'customer_id', 
            x = 'Total_amount_spent',  
            palette =colors[::-1],
            hue = 'customer_id',
           )

#Label Axes
plt.xlabel("Total Ammount Spent", fontweight= 'bold', fontsize= 14)
plt.ylabel("Customer_id", fontsize = 14, fontweight = 'bold')
plt.title("Premium Customer Spendings")
plt.tight_layout()
plt.savefig("Premium_Customers.png", dpi = 300, bbox_inches = 'tight')
plt.show()
```
