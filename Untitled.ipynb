```python
# Import Important Libraries
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import pyodbc 
import os
```

```python
# Define Connection Parameters
server = 'DESKTOP-U507UFU'
database = 'Practice'
#define Connection String
conn_str = (
    "Driver={SQL Server};"
    f"Server={server};"
    f"Database={database};"
    "Trusted_Connection=yes;"   
)
 #Establish Connection
conn = pyodbc.connect(conn_str)
```

```python
#Run Query to get data into dataframe
query = """
        WITH order_cte AS(
        SELECT order_id, item,
        LEAD(item) OVER (ORDER BY order_id) AS next_item,
        LAG(item) OVER (ORDER BY order_id) AS prev_item
        FROM complaints       
        )
    SELECT order_id AS corrected_order_id,
        CASE WHEN
        (order_id % 2) = 1 AND next_item IS NOT NULL 
        THEN next_item
        WHEN (order_id % 2) = 0 THEN prev_item
        ELSE item 
        END AS item 
    FROM order_cte 
    ORDER BY corrected_order_id;
        """
df = pd.read_sql(query, conn)
    
```

```python
#Close Connection
conn.close()
```
