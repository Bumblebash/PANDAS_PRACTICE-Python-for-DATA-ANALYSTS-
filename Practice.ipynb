# Sort data 

```python
# Remove White spaces 
df.columns = df.columns.str.strip()
```

```python
df = df.sort_values( by= ['Total_ordered'], ascending = False)
```

```python
#Top10
top10 = df.head(10)
bottom10 = df.tail(10)
```

```python
#Visualize 
fig,axes = plt.subplots(1,2, figsize=(16,6))
axes[0].barh(top10['Book_title'], top10['Total_ordered'], color = 'green')
axes[0].set_ylabel("Book_tiltle", fontsize=14, fontweight = 'bold')
axes[0].set_xlabel("Total_ordered", fontsize=14, fontweight='bold')
axes[0].invert_yaxis()
axes[0].set_title("Best selling Books")
#Second plot
axes[1].barh(bottom10['Book_title'], bottom10['Total_ordered'], color = 'green')
axes[1].set_ylabel("Book_tiltle", fontsize=14, fontweight = 'bold')
axes[1].set_xlabel("Total_ordered", fontsize=14, fontweight='bold')
axes[1].invert_yaxis()
axes[1].set_title("Worst Selling Books")
fig.suptitle("Sales Perfomance", fontweight = 'bold')
plt.savefig('Sales_perfomance.png', dpi = 300, bbox_inches = 'tight')
plt.tight_layout()
```
